```python
import pandas as pd
import matplotlib.pyplot as plt

# Read the CSV files
vit_df = pd.read_csv('./vit_training_metrics.csv')
moe_df = pd.read_csv('./moe_vit_metrics.csv')
swin_df = pd.read_csv('./swin_transformer_metric.csv')
nat_df = pd.read_csv('./nat_training_metrics.csv')


# Create figure with 2x2 subplots
fig, ((ax1, ax2), (ax3, ax4)) = plt.subplots(2, 2, figsize=(15, 12))

# Create epochs array
epochs = range(1, 21)

# Plot training accuracy
ax1.plot(epochs, vit_df['train_accuracy']*100, label='ViT')
ax1.plot(epochs, moe_df['train_accuracy'], label='MoE-ViT')
ax1.plot(epochs, swin_df['Train Accuracy']*100, label='Swin')
ax1.plot(epochs, nat_df['train_accuracy']*100, label='NAT')
ax1.set_title('Training Accuracy')
ax1.set_xlabel('Epoch')
ax1.set_ylabel('Accuracy (%)')
ax1.legend()
ax1.grid(True)

# Plot validation accuracy
ax2.plot(epochs, vit_df['valid_accuracy']*100, label='ViT')
ax2.plot(epochs, moe_df['test_accuracy'], label='MoE-ViT')
ax2.plot(epochs, swin_df['Valid Accuracy']*100, label='Swin')
ax2.plot(epochs, nat_df['valid_accuracy']*100, label='NAT')
ax2.set_title('Test/Validation Accuracy')
ax2.set_xlabel('Epoch')
ax2.set_ylabel('Accuracy (%)')
ax2.legend()
ax2.grid(True)

# Plot training loss
ax3.plot(epochs, vit_df['train_loss'], label='ViT')
ax3.plot(epochs, moe_df['train_loss'], label='MoE-ViT')
ax3.plot(epochs, swin_df['Train Loss'], label='Swin')
ax3.set_title('Training Loss')
ax3.set_xlabel('Epoch')
ax3.set_ylabel('Loss')
ax3.legend()
ax3.grid(True)

# Plot validation loss
ax4.plot(epochs, vit_df['valid_loss'], label='ViT')
ax4.plot(epochs, moe_df['test_loss'], label='MoE-ViT')
ax4.plot(epochs, swin_df['Valid Loss'], label='Swin')
ax4.set_title('Test/Validation Loss')
ax4.set_xlabel('Epoch')
ax4.set_ylabel('Loss')
ax4.legend()
ax4.grid(True)

plt.tight_layout()
plt.show()
```
